```python
#Functions
def sum(lb,ub):
    total=0
    while(lb<=ub):
        total+=lb
        lb+=1
    return total

sum(1,20)
```

## Lambda Funtion

```python
#id 
def id(lb): 
    return lb

#square 
def sqr(lb): 
    return lb*lb

#cube 
def cube(lb): 
    return lb*lb*lb

def doitall(func,lb,ub):
    total=0 
    while(lb<=ub): 
        total+=func(lb) 
        lb+=1 
    return total

#Sum of Numbers
#doitall(id,1,10)

#SUm of Squares 
#doitall(sqr,1,10)

#sum of cubes 
doitall(cube,1,10)
```
