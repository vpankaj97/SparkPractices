## RDD APIs

```python
#Printing the first element of RDD
print(orderItems.first())
print(readme.first())
```

```python
#Counting number of items in RDD
print(orderItems.count())
print(readme.count())
```

```python
#Looping using Take function
for i in orderItems.take(10):
    print(i)
```
